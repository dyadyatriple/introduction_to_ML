```python
import pandas as pd
from sklearn.model_selection import GridSearchCV,KFold,cross_val_score
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import scale

df=pd.read_csv('./wine.data',
               names= ["class", "alcohol", "malic_acid", "ash", "ash_alcalinity", "magnesium", "total_phenols", "flavanoids", 
    "nonflavanoid_phenols", "proanthocyanins", "color_intensity", "hue", "od280", "proline"])
```

```python
def save_answer(filename,answer):
    with open(filename+'.txt','w') as f:
        f.write(str(answer))
```

```python
df
```

```python
def best_neighb_scores(X,y):
    cv=KFold(n_splits=5,shuffle=True,random_state=42,)
    clf=KNeighborsClassifier(n_neighbors=1)
    gs=GridSearchCV(estimator=clf,
                    param_grid={'n_neighbors':range(1,51)},
                   cv=cv,
                   scoring='accuracy')
    gs.fit(X,y)
    scores=cross_val_score(estimator=gs.best_estimator_,X=X,y=y,cv=cv,scoring='accuracy')
    return scores.mean(),gs.best_params_['n_neighbors']
```

```python
best_score,opt_k=best_neighb_scores(df.drop('class',axis=1),df['class'])
```

```python
save_answer('1',opt_k)
save_answer('2',round(best_score,2))
print('The best n of neighbours is {} and the best score is {}'.format(best_score,opt_k))
```

```python
X=df.drop('class',axis=1)
scaled=scale(X.values)
```

```python
scaled.shape
```

```python
best_score,opt_k=best_neighb_scores(scaled,df['class'])
```

```python
save_answer('3',opt_k)
save_answer('4',round(best_score,2))
print('After Scaling the best n of neighbours is {} and the best score is {}'.format(best_score,opt_k))
```
